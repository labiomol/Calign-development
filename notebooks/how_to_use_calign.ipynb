## A simple guide to using Calign.

```python
#Import the tool and its key function for use.
from calign import align_epitopes
```

```python
# Here we create a variable called "resultado" and call the function "align_epitopes",
#passing as arguments the names of the files needed to perform the analysis
#(the paths to the folders leading to the files can also be used).

result = align_epitopes(
    "proteina.fasta",
    "epitopos.fasta"
)
```

## After executing the command above, you will be able to find the results in the folder where your files are located.

```python
from IPython.display import Image
Image(filename=result['output_png'])
```
